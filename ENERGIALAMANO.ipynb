```python
#columna_a = df['Refrigerator'] + df['Clothes washer'] + df['Clothes Iron'] + df['Computer'] + df['Oven'] + df['Play'] + df['TV'] + df['Sound system']

df['fantasma'] = df['Medidor [W]'] - (df['Refrigerator'] + df['Clothes washer'] + df['Clothes Iron'] + df['Computer'] + df['Oven'] + df['Play'] + df['TV'] + df['Sound system'])
df['Medidor Limpio [W]'] = df['Medidor [W]'] - df['fantasma']

df = df.drop(['Medidor [W]', 'fantasma'], axis=1)


df.to_csv('datoslimpios.csv', index=False)
```

```python
df = pd.read_csv('datoslimpios.csv')

df['coste total'] = df['Medidor Limpio [W]'] * 0.9
df['refrigerador costo'] = df['Refrigerator'] * 0.9
df['lavadora costo'] = df['Clothes washer'] * 0.9
df['plancha costo'] = df['Clothes Iron'] * 0.9
df['computadora costo'] = df['Computer'] * 0.9
df['horno costo'] = df['Oven'] * 0.9
df['play costo'] = df['Play'] * 0.9
df['TV costo'] = df['TV'] * 0.9
df['sistema de sonido costo'] = df['Sound system'] * 0.9

df = df.drop(['Refrigerator', 'Clothes washer', 'Clothes Iron', 'Computer', 'Oven', 'Play', 'TV', 'Sound system'], axis=1)

df.to_csv('costos.csv', index=False)
```

```python
df = pd.read_csv('costos.csv')

mes = 43800

df['group'] = np.arange(len(df)) // mes
df['Coste Total X Mes'] = df.groupby('group')['coste total'].sum()
df['Coste Refrigerador X Mes'] = df.groupby('group')['refrigerador costo'].sum()
df['Coste Lavadora X Mes'] = df.groupby('group')['lavadora costo'].sum()
df['Coste Plancha X Mes'] = df.groupby('group')['plancha costo'].sum()
df['Coste Computadora X Mes'] = df.groupby('group')['computadora costo'].sum()
df['Coste Horno X Mes'] = df.groupby('group')['horno costo'].sum()
df['Coste Play X Mes'] = df.groupby('group')['play costo'].sum()
df['Coste TV X Mes'] = df.groupby('group')['TV costo'].sum()
df['Coste Sistema de Sonido X Mes'] = df.groupby('group')['sistema de sonido costo'].sum()

df = df.drop(['coste total', 'refrigerador costo', 'lavadora costo', 'plancha costo', 'computadora costo', 'horno costo', 'play costo', 'TV costo', 'sistema de sonido costo', 'id', 'Fecha', 'Medidor Limpio [W]', 'group'], axis=1)
df.to_csv('costestotal.csv', index=False)
```

```python
df = pd.read_csv('costos.csv')

mes = 43800

df['group'] = np.arange(len(df)) // mes
df['promedio Coste x Minuto'] = df.groupby('group')['coste total'].sum()/mes
df['Promedio Coste Refrigerador x Minuto'] = df.groupby('group')['refrigerador costo'].sum()/mes
df['Promedio Coste Lavadora X Minuto'] = df.groupby('group')['lavadora costo'].sum()/mes
df['Promedio Coste Plancha X Minuto'] = df.groupby('group')['plancha costo'].sum()/mes
df['Promedio Coste Computadora X Minuto'] = df.groupby('group')['computadora costo'].sum()/mes
df['Promedio Coste Horno X Minuto'] = df.groupby('group')['horno costo'].sum()/mes
df['Promedio Coste Play X Minuto'] = df.groupby('group')['play costo'].sum()/mes
df['Promedio Coste TV X Minuto'] = df.groupby('group')['TV costo'].sum()/mes
df['Promedio Coste Sistema De Sonido X Minuto'] = df.groupby('group')['sistema de sonido costo'].sum()/mes

df = df.drop(['coste total', 'refrigerador costo', 'lavadora costo', 'plancha costo', 'computadora costo', 'horno costo', 'play costo', 'TV costo', 'sistema de sonido costo', 'id', 'Fecha', 'Medidor Limpio [W]', 'group'], axis=1)
df.to_csv('promedioxminuto.csv', index=False)
```
